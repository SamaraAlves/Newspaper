# Extração de News

```python
# Bibliotecas
import newspaper
from newspaper import Article
import pandas as pd
from pandas import ExcelWriter
from openpyxl.workbook import Workbook
import datetime

```

```python
#Lista de sites
https_ = {
    'cbc': 'http://www.cbc.ca/',
    'nanowerk': 'https://www.nanowerk.com/',
    'phys': 'https://phys.org/',
    '3dprintingprogress': 'https://www.3dprintingprogress.com/',
    'digitaltrends': 'https://www.digitaltrends.com/',
    'sky': 'https://news.sky.com/',
    'smithsonianmag': 'https://www.smithsonianmag.com/',
    'popularmechanics': 'https://www.popularmechanics.com/',
    'the_scientist': 'https://www.the-scientist.com/',
    'sciencealert': 'https://www.sciencealert.com/',
    'sciencemag': 'http://www.sciencemag.org/',
    'fastcompany': 'https://www.fastcompany.com/',
    'medium': 'https://medium.com/',
    'reuters': 'https://www.reuters.com/',
    'forbes': 'https://www.forbes.com/',
    'bloomberg': 'https://www.bloomberg.com/',
    'webnewswire': 'http://webnewswire.com/',
    'wesrch': 'https://www.wesrch.com/',
    'beforeitsnews': 'http://beforeitsnews.com/',
    'cnn': 'https://cnn.com',
    'bbc': 'https://www.bbc.com/news',
    'nature': 'https://www.nature.com/news',
    'wired': 'https://www.wired.com/category/science/',
    'newscientist': 'https://www.newscientist.com/',
    'livescience': 'https://www.livescience.com/'
    }
```

```python
def extract_url(https_):
    '''Seleciona as paginas de cada um das companhias'''
    company = list(https_.keys())
    url = list(https_.values())

    url_ = []
    for i in range(2):
        #Coletar artigos disponíveis com a url
        print ('Download artigo companhia:', company[i])
        paper_ = newspaper.build(str(url[i]), language='en')
        for article in paper_.articles:
            url_.append(article.url)
    return url_        
```

```python
def extract_info(url_):
    '''Extrai Title, Text, Author, Publish Data e PageLinke cria um dicionario de informacoes'''
    news = {}
    news['title'] = []
    news['text'] = []
    news['author'] = []
    news['publish_date'] = []
    news['pagelink'] = []

    for i in range(len(url_)):
        try:
            article = Article(str((url_[i])))
            article.download()
            article.parse()
            news['title'].append(article.title)
            news['text'].append(article.text)
            news['author'].append(article.authors)
            news['publish_date'].append(article.publish_date)
            news['pagelink'].append(url_[i])
        except:
            print ('Não foi possivel fazer o download:', url_[i])
            pass
    return news
```

```python
def date_news(news):
    '''se nao temos a data extraída direto do site colocamos a data do dia em que foi feita a coleta de news'''
    date_ = []
    for i in range(len(news['publish_date'])):
        if news['publish_date'][i] == None:
            today_date = datetime.datetime.today().strftime('%Y-%m-%d')
            date_.append(today_date)
        else:     
            date_.append(news['publish_date'][i])

    news['date'] = date_ 
    return news['date']
```

```python
url = extract_url(https_)
```

```python
news = extract_info(url)
```

```python
news['date'] = date_news(news)
```

```python
#Transforma arquivo em pandas dataframe
news = pd.DataFrame(data=news)
news.drop('publish_date', axis = 1, inplace = True)

display(news)
```

```python
#Salvar Arquivo
writer = ExcelWriter('News.xlsx')
news.to_excel(writer,'Data')
writer.save()
```
